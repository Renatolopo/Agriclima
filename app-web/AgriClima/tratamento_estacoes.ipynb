```python
ana = estacoes[['Nome', 'Codigo', 'TipoEstacao', 'Latitude', 'Longitude']]
ana['Fonte'] = 'ANA'
```

```python
ana
```

```python
inmet_automaticos = pd.read_csv(r"C:\Users\renat\Downloads\CatalogoEstaçõesAutomáticas.csv", delimiter=';')
inmet_automaticos['TipoEstacao'] = 'Automatica'
inmet_automaticos['Fonte'] = 'INMET'
inmet_automaticos = inmet_automaticos[['DC_NOME', 'CD_ESTACAO', 'TipoEstacao', 'VL_LATITUDE', 'VL_LONGITUDE', 'Fonte']]
inmet_automaticos.columns = ['Nome', 'Codigo', 'TipoEstacao', 'Latitude', 'Longitude', 'Fonte']
inmet_automaticos
```

```python
inmet_automaticos[inmet_automaticos['Nome'] == 'JANUARIA']
```

```python
inmet_convencionais = pd.read_csv(r"C:\Users\renat\Downloads\CatalogoEstaçõesConvencionais.csv", delimiter=';')
inmet_convencionais['TipoEstacao'] = 'Convencionais'
inmet_convencionais['Fonte'] = 'INMET'
inmet_convencionais = inmet_convencionais[['DC_NOME', 'CD_ESTACAO', 'TipoEstacao', 'VL_LATITUDE', 'VL_LONGITUDE', 'Fonte']]
inmet_convencionais.columns = ['Nome', 'Codigo', 'TipoEstacao', 'Latitude', 'Longitude', 'Fonte']
inmet_convencionais
```

```python
estacoes_tratadas = pd.concat([inmet_automaticos, inmet_convencionais, ana])
```

```python
estacoes_tratadas
```

```python
estacoes_tratadas.to_csv('../../data/Estacao_tratada.csv', index='False')
```
